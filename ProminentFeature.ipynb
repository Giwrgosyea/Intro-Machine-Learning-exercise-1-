```python
import numpy as np
from sklearn import linear_model
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from numpy import *
from matplotlib.pyplot import *
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
data = pd.read_csv(dataset_url, sep=';')

#min_max_scaler = preprocessing.MinMaxScaler()
#balance_data = min_max_scaler.fit_transform(balance_data)

X = data.alcohol.values
X= X.reshape(len(X),1)

Y = data.quality.values
Y = Y.reshape(len(Y),1)

X_train = X[:-20]
X_test = X[-20:]
y_train = Y[:-20]
y_test = Y[-20:]

epoch=1
plt.scatter(X ,Y,  color='black')
plt.show()
list1=[]
list2=[]
list3=[]
#min_max_scaler = preprocessing.MinMaxScaler()
#X = min_max_scaler.fit_transform(X)
while epoch <=100:
    
    clf = linear_model.SGDRegressor(loss='squared_loss',fit_intercept=True, shuffle=True,learning_rate='invscaling',eta0=0.001)
    clf.fit(X_train, y_train)
    
  
    list1.append(clf.coef_)
    
  
    list2.append(clf.intercept_)
    
    y_pred = clf.predict(X_test)
    
    ms = mean_squared_error(y_test, y_pred)
    list3.append(ms)
    
    epoch = epoch + 1


fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(list1,list2,list3)
pyplot.show()
```
