```python
import numpy as np
import scipy as sc 
from sklearn import linear_model
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error,log_loss
from numpy import *
from matplotlib.pyplot import *
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import least_squares
```

```python
dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
data = pd.read_csv(dataset_url, sep=';')
```

```python
# For each one of the prominent features selected in step a)

X = data.alcohol.values
X= X.reshape(-1,1)

Y = data.quality.values
Y = Y.reshape(-1,1)
```

```python
def fun(x):
    return mean_squared_error(x[0],x[1])
```

```python
#split the data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2) 
#clf = linear_model.LogisticRegression()]
epoch=1
list1=[]
list2=[]
list3=[]
list4 = []
while epoch <=100:
    clf = linear_model.SGDClassifier(alpha=0.01, average=False, class_weight=None, epsilon=0.1,
           eta0=0.01, fit_intercept=True, l1_ratio=0.15,
           learning_rate='optimal', loss='log', n_iter=epoch,
           n_jobs=1, penalty='l2', power_t=0.5, random_state=None,
           shuffle=True,verbose=0, warm_start=False)
    clf.fit(X_train, y_train)
    y_pred= clf.predict(X_test)
    
    list1.append(np.sum(clf.coef_)/len(clf.coef_))
    list2.append(np.sum(clf.intercept_)/len(clf.intercept_))
    cost= mean_squared_error(y_test, y_pred)
    pred = clf.predict_proba(X_test)
    log_lost=log_loss(y_test,pred)
    list4.append(log_lost)
    list3.append(cost)

    epoch=epoch+1

##plot cost with epochs 




fig1 = pyplot.figure()
ax1 = Axes3D(fig1)
ax1.scatter(list1,list2,list4)
plt.show()

```

```python

fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(list1,list2,list3)
plt.show()
```
